```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv("/Users/hamsinisankaran/Desktop/logisticregression/Social_Network_Ads.csv")
```

```python
X = dataset.iloc[:,  [2,3]].values
y = dataset.iloc[:, 4].values
```

```python
X 
```

```python
y
```

```python
#splitting the dataset into training and test dataset 
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state  = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
#feature scaling . Preprocessing the dataset. Bring the X_train and X_test into a scale , 
#such that there is 0 variance 
#ytrain and ytest need not be scaled 
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#fitting logistic regression to the training . 
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)
```

```python
#predicting the results of Logistic Regression with the test data 
y_predict = classifier.predict(X_test)
```

```python
y_predict
```

```python
X_test
```

```python
#evaluation of Logistic Regression model 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
```

```python
cm
```

```python
#visualization of the results 
#red points did not buy SUV, green points bought SUV 
# from observation, guys who had the lowest median salary purchased SUV 
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Classifier (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
